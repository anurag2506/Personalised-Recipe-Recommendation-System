```python
def getJson(item):
    chat_completion = client.chat.completions.create(

    messages=[
        {
            "role": "system",
            "content": "You are an expert Chef who has worked all over the world. For a food item mentioned in the prompt, I want you to mention the following details as a json"
            " The output should have only the json for a particular item and not any other extra text not even a extra single char, just the json : "
            "1. Dietary Category:(veg, non-veg, keto, paleo etc.). MAKE SURE THAT THE CASE OF THE KEY IS SAME AND THE VALUE IS NON EMPTY"
            "2. Specific Cuisine: (Italian, Indian ,Chinese etc.). MAKE SURE THAT THE CASE OF THE KEY IS SAME AND THE VALUE IS NON EMPTY"
            "3. Ingredients: vegetables,spices,etc.(python list). MAKE SURE THAT THE CASE OF THE KEY IS SAME AND THE VALUE IS NON EMPTY"
            "4. Preparation Time: 25, 30, 40 . MAKE SURE THAT THE CASE OF THE KEY IS SAME AND THE VALUE IS NON EMPTY AND JUST THE INTEGER AND NOT ANY OTHER CHARACTER OR STRING"
            "USE ONLY THE KEYS MENTIONED IN THE SAME CASE AND DON'T CHANGE THE KEY AT ALL"
            "Make sure that the json is formatted and well written without any extra characters and just the key value pairs"
            "Andt the KEYS IN ALL THE JSONS SHOULD BE SAME AS MENTIONED AND CONSISTENT SO THAT IT BECOMES EASY TO QUERY A KEY . SAME KEYS REMEMBER"
        },
        # Set a user message for the assistant to respond to.
        {
            "role": "user",
            "content": f"{item}",
        }
    ],

    # The language model which will generate the completion.
    model="llama3-70b-8192",

    #
    # Optional parameters
    #

    # Controls randomness: lowering results in less random completions.
    # As the temperature approaches zero, the model will become deterministic
    # and repetitive.
    temperature=0.5,

    # The maximum number of tokens to generate. Requests can use up to
    # 2048 tokens shared between prompt and completion.
    max_tokens=1024,

    # Controls diversity via nucleus sampling: 0.5 means half of all
    # likelihood-weighted options are considered.
    top_p=1,

    # A stop sequence is a predefined or user-specified text string that
    # signals an AI to stop generating content, ensuring its responses
    # remain focused and concise. Examples include punctuation marks and
    # markers like "[end]".
    stop=None,

    # If set, partial message deltas will be sent.
    stream=False
)
    return chat_completion.choices[0].message.content
```

```python
info_json = []

for item in items:
    response = getJson(item)
    info_json.append(response)
    
```
